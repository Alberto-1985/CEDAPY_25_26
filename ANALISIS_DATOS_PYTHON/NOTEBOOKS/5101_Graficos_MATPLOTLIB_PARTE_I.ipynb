# <a id='toc1_'></a>[Creación de gráficos con Matplotlib - Parte I](#toc0_)

Para crear un gráfico con **matplotlib** es común seguir los siguientes pasos:

- Importar el módulo **pyplot**.
- Definir la figura que contendrá el gráfico, que es la region (ventana o página) donde se dibujará y los ejes sobre los que se dibujarán los datos. Para ello se utiliza la función *subplots()*.
- Dibujar los datos sobre los ejes. Para ello se utilizan distintas funciones dependiendo del tipo de gráfico que se quiera.
- Personalizar el gráfico. Para ello existen multitud de funciones que permiten añadir un título, una leyenda, una rejilla, cambiar colores o personalizar los ejes.
- Guardar el gráfico. Para ello se utiliza la función *savefig()*.
- Mostrar el gráfico. Para ello se utiliza la función *show()*.

````python
# Importar el módulo pyplot con el alias plt
import matplotlib.pyplot as plt
# Crear la figura y los ejes
fig, ax = plt.subplots()
# Dibujar puntos
ax.scatter(x = [1, 2, 3], y = [3, 2, 1])
# Guardar el gráfico en formato png
plt.savefig('diagrama-dispersion.png')
# Mostrar el gráfico
plt.show()
````

Vamos a utilizar un dataset con información sobre fútbol de las ligas más importantes en Europa:

````python
df.info()
````

````python
# Filtrando datos del dataset entre 2014 y 2019
messi_goals = df.loc['Lionel Messi'].query('year != 2020')
ronaldo_goals = df.loc['Cristiano Ronaldo'].query('year != 2020')

# Plot -> ax.plot(X, Y)
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals)
ax.plot(ronaldo_goals.year, ronaldo_goals.goals)
# plt.savefig('1.jpg')
plt.show()
````

### <a id='toc1_1_2_'></a>[Preparando un diagrama (plot)](#toc0_)

Hay una serie de argumentos que permiten personalizar cómo se ven los datos de un gráfico tanto como se pueda imaginar:

- **marker**='o' → añade marcadores al plot y también puedes indicar el tipo de marcador que quieres. 
[Aquí](https://matplotlib.org/stable/api/markers_api.html) figuran todas las opciones de marcadores que hay.   

- **linestyle**='--' → cambia el tipo de línea. Así mismo, [aquí](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) se pueden encontrar los distintos tipos de línea. Incluso si se coloca 'None', simplemente ya no habrá líneas y quedarán solo los marcadores.
- **color**='b' → cambia el color de los datos. [Aquí](https://matplotlib.org/stable/tutorials/colors/colors.html) se encuentran cómo poner cualquier color e incluso dejarlos transparentes.
- **label**='name' → pone una etiqueta al plot (muy útil cuando se quiere poner varias gráficas en un solo plot).    


Adicionalmente, el objeto **Axes** trae una serie de métodos que se pueden usar para que sea sencillo identificar de qué trata la gráfica. Se deben siempre antes de mostrar la gráfica con ```plt.show()```.
```
.set_xlabel('Name') → coloca nombre al eje de las X.
.set_ylabel('Name') → coloca nombre al eje de las Y.
.set_title('Name') → coloca el título de tu plot.
.set(xlabel='Name') → también puedes usar únicamente set y pasar como parámetros lo anterior.
.legend() → si pusiste etiquetas, legend las mostrará.
```

````python
fig, ax = plt.subplots()
ax.plot(messi_goals.year, messi_goals.goals,
        marker='o', color='b', linestyle='--', label='Lionel Messi')
ax.plot(ronaldo_goals.year, ronaldo_goals.goals,
        marker='s', color='r', linestyle='dotted', label='Cristiano Ronaldo')

ax.set(xlabel='Años', ylabel='Goles', title='Messi vs. Cristiano - GOLES (2014 - 2019)')
ax.legend()
# plt.savefig('2.jpg')
plt.show()
````

## <a id='toc1_3_'></a>[Diagramas de dispersión o puntos](#toc0_)

Un gráfico de dispersión es la opción adecuada para visualizar la relación entre dos conjuntos de datos emparejados.   

**scatter(x, y)**: Dibuja un diagrama de puntos con las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y. 

````python
import matplotlib.pyplot as plt
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes) # eje x friends, eje y minutes
# etiqueta cada punto
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
    xy=(friend_count, minute_count),
    xytext=(5, -5),
    textcoords='offset points')
# Pone la etiqueta con su punto
# pero un poco desplazada
plt.title("Minutos diarios vs. Nº de amigos")
plt.xlabel("Nº de amigos")
plt.ylabel("Minutos diarios en la red social")
plt.show()
````

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
plt.scatter(test_1_grades, test_2_grades)
plt.title("Ejes no comparables por diferente escala")
plt.xlabel("test 1")
plt.ylabel("test 2")
plt.show()
````

````python
# Igualamos las escalas de los ejes
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
plt.scatter(test_1_grades, test_2_grades)
plt.title("Ejes comparables igualando las escalas")
plt.xlabel("test 1")
plt.ylabel("test 2")
plt.axis("equal")  # Igualar las escalas de los ejes
plt.show()
````

El tercer argumento de **plt.bar** especifica la anchura de las barras. Hemos elegido una anchura de 10, para llenar así todo el decil. También hemos desplazado las barras a la derecha en 5, de forma que, por ejemplo, la barra “10” (que corresponde al decil 10-20) tendría su cen ro en 15 y, por lo tanto, ocuparía el rango correcto.       

También añadimos un borde negro a cada barra para distinguirlas de forma visual. La llamada a **plt.axis** indica que queremos que el eje x vaya desde -5 hasta 105 (solo para dejar un poco de espacio a la izquierda y a la derecha), y que el eje y varíe de 0 a 5; la llamada a plt.xticks pone las etiquetas del eje x en 0, 10, 20, ..., 100.   

Conviene ser juiciosos al utilizar *plt.axis*. Cuando se crean gráficos de barras, está especialmente mal considerado que el eje y **no empiece en 0**, ya que de ese modo la gente se confunde con mucha facilidad.

````python
import matplotlib.pyplot as plt
mentions = [500, 505]
years = [2017, 2018]
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("Nº de veces que he oído a alguien decir «ciencia de datos»")
# si no se hace esto, matplotlib etiquetará el eje x con 0, 1
# y añadirá +2.013e3 en la esquina (¡qué malo es matplotlib!)
plt.ticklabel_format(useOffset=False)
# el eje y erróneo solo muestra la parte sobre 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("¡Mira el «enorme» aumento!")
plt.show()
````

## <a id='toc1_8_'></a>[Histogramas](#toc0_)

Los histogramas muestran ***la forma de sus datos***. El eje horizontal muestra sus valores de datos, con cada barra correspondiendo a un rango de valores. El eje vertical muestra cuántos puntos de datos tienen valores en el rango de cada barra. 

**hist(x, bins)**: Dibuja un histograma con las frecuencias resultantes de agrupar los datos de la lista x en las clases definidas por la lista **bins**.   


Se puede personalizar el gráfico con otros argumentos:

- ```bins=n``` → que, como ya se ha visto, indica cuántas barras serán mostradas en el histograma. Por defecto son 10.
- A este argumento también se le puede pasar una lista para establecer un límite en el rango de las X en las que se quiere que se muestren.
- Si se quiere colocar 2 histogramas que se podrían sobreponer entre sí, se puede usar el argumento **alpha** que si le pasa un valor entre 0 y 1 cambia la transparencia del color.
- También se puede usar el argumento ```histtype='step'``` que hace que cambie la forma que se representa el histograma, por ejemplo, líneas en vez de rectángulos sólidos. [Aquí](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) está la documentación para ver qué otros tipos de histtype existen y también otros argumentos de hist.
- Por último, si se usa ```density=True```, el gráfico se trasformará a porcentajes para que el área total de los rectángulos sea de 1 (100%). Se recomienda recordar este argumento porque podría ser muy útil más adelante para entender el peso de los valores, más que su cantidad en sí misma.


````python
# Usando el dataset de goles por equipo y año
# Goles totales por equipo en 2018
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()
#print(team_goals_2018)

plt.figure(figsize=[7, 6]) # 7 in de ancho y 6 in de alto
plt.hist(team_goals_2018, bins=25)
plt.title('Goles totales por equipo en 2018')
plt.xlabel('Cantidad de goles')
plt.ylabel('Número de equipos con X cantidad de goles')
plt.show()
````

````python
# Goles totales por equipo en 2018 - Formato "step" -> solo silueta
df_2018 = df.query('year == 2018')
team_goals_2018 = df_2018.groupby('team_name')['goals'].sum()
plt.hist(team_goals_2018, bins=15, histtype='step', density=True) # step=silueta
plt.show()
````

## <a id='toc1_10_'></a>[Diagramas de caja y bigotes](#toc0_)


Los diagramas de caja ayudan a ver el centro y la extensión de los datos. También se pueden utilizar como herramienta visual para comprobar normalidad o identificar puntos que podrían ser valores atípicos.   

Estas son las partes básicas de un diagrama de caja:

- La línea central de la caja indica la mediana de los datos. Una mitad de los datos está por debajo de este valor, y la otra por encima. Si los datos son simétricos, la mediana estará en el centro de la caja. Si los datos están sesgados, la mediana estará más cerca de la parte superior o inferior de la caja.
- Los extremos de arriba y abajo de la caja indican los cuantiles, o percentiles, 25 y 75. Estos dos cuantiles también se conocen como cuartiles, porque separan cuartos (25 %) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartílico (IQR).
- Las líneas que se extienden desde la caja se llaman bigotes. Los bigotes representan la varianza esperada de los datos. Estos bigotes se extienden 1,5 veces el IQR desde los extremos superior e inferior de la caja. Si los datos no llegan hasta el final de los bigotes, estos se ajustan a los valores mínimo y máximo de los datos. Si hay datos que queden por encima o por debajo de los extremos de los bigotes, se los representa con puntos. Estos puntos se conocen como valores atípicos. Un valor atípico es el que supera la varianza esperada. Merece la pena revisar estos puntos de datos para aclarar si son atípicos o erróneos. Los bigotes no incluyen dichos valores.


**boxplot(x)**: Dibuja un diagrama de caja y bigotes con los datos de la lista **x**. 

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.boxplot([1, 2, 1, 2, 3, 4, 3, 3, 5, 7])
plt.show()
````

## <a id='toc1_12_'></a>[Diagramas de contorno](#toc0_)

Los diagramas o gráficos de contorno muestran la relación tridimensional en dos dimensiones, graficando los factores X y Y (predictores) en las escalas X y Y y los valores de respuesta se representan mediante contornos. Un gráfico de contorno es como un mapa topográfico en el que se grafican valores de X, Y y Z, en lugar de longitud, latitud y elevación.

**contourf(x, y, z)**: Dibuja un diagrama de contorno con las curvas de nivel de la superficie dada por los puntos con las coordenadas de las listas **x**, **y** y **z** en los ejes X, Y y Z respectivamente. 

````python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
````

## <a id='toc1_13_'></a>[Mapas de calor](#toc0_)

Los mapas de calor son una popular técnica de visualización de datos que utiliza el color para representar diferentes niveles de magnitud de los datos, lo que permite identificar rápidamente patrones y anomalías en el conjunto de datos.   




**imshow(x)**: Dibuja un mapa de color a partir de una matriz (array bidimensiona) **x**. 


````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.random.random((16, 16))
ax.imshow(x)
plt.show()
````

## <a id='toc1_14_'></a>[Diagrama de Áreas Apiladas (stackplot())](#toc0_)

Se representa la evolución acumulada de tres categorías a lo largo del tiempo.   

Cuando se trata de conjuntos de datos compuestos por múltiples categorías o componentes, el gráfico de áreas apiladas se convierte en un activo valioso. En esta combinación, los segmentos de áreas individuales se apilan uno sobre otro, culminando en una forma compuesta que retrata tanto la tendencia general como la contribución de cada categoría a esa tendencia.

El gráfico de áreas apiladas facilita la comparación de los componentes de los datos manteniendo el enfoque en el conjunto acumulativo. Esta combinación es particularmente eficaz para ilustrar la distribución y las proporciones de los datos dentro de un contexto más amplio, aportando información sobre cómo las diferentes categorías contribuyen al patrón general.

**plt.stackplot(x, y1, y2, ..., labels=['Etiqueta1', 'Etiqueta2', ...])**


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y1 = np.array([1, 3, 4, 8, 10])
y2 = np.array([2, 2, 3, 3, 4])
y3 = np.array([1, 2, 2, 3, 5])

plt.stackplot(x, y1, y2, y3, labels=['Categoría A', 'Categoría B', 'Categoría C'])
plt.legend()
plt.title('Evolución de categorías en el tiempo')
plt.xlabel('Tiempo')
plt.ylabel('Valor acumulado')
plt.show()
````

## <a id='toc1_17_'></a>[Gráfico de barras apiladas](#toc0_)

Los gráficos de barras apiladas representan los valores que toman ciertos grupos dentro de otros grupos.    
Por ejemplo, si se considera que personas que viven en cinco ciudades diferentes (esto es, los grupos 'G1', 'G2', 'G3', 'G4' y 'G5') respondieron Si o No a una pregunta y los resultados (el conteo de respuestas en cada ciudad) se guardaron en las variables valores1 y valores2.

Estos datos se pueden representar en Python con la función bar de matplotlib haciendo uso del argumento **bottom**, que representa los valores del eje Y en los que dibujar la parte inferior de las barras, lo que permite apilar unas barras sobre otras.

````python
import matplotlib.pyplot as plt

# Datos
grupos = ['G1', 'G2', 'G3', 'G4', 'G5']
valores1 = [12, 19, 14, 27, 16]
valores2 = [21, 30, 15, 17, 20]

fig, ax = plt.subplots()

# Gráfico de barras apiladas
ax.bar(grupos, valores1)
ax.bar(grupos, valores2, bottom = valores1) #valores 1 esta debajo 

# plt.show() 
 
````

### <a id='toc1_17_1_'></a>[Generalización para varios subgrupos](#toc0_)

Si no se quiere apilar los valores manualmente, se puede unir los valores en un array de numpy y luego utilizar el siguiente bucle for:

````python
import matplotlib.pyplot as plt
import numpy as np

# Datos
grupos = ['G1', 'G2', 'G3', 'G4', 'G5']

values = np.array([[12, 19, 14, 27, 16],
                  [21, 30, 15, 17, 20],
                  [15, 23, 12, 11, 15],
                  [2, 5, 1, 6, 8]])

fig, ax = plt.subplots()

# Gráfico de barras apiladas con un bucle
for i in range(values.shape[0]):
  ax.bar(grupos, values[i], bottom = np.sum(values[:i], axis = 0))

# plt.show() 
````

## <a id='toc1_18_'></a>[Gráfico de Vectores (quiver())](#toc0_)

Se visualizan vectores en un espacio bidimensional, con flechas indicando dirección y magnitud.   
Se usa para representar direcciones y magnitudes de fuerzas o velocidades en un espacio bidimensional.

**plt.quiver(X, Y, U, V)**

Parámetros
- X, Y	--> Coordenadas de los puntos donde se dibujarán los vectores. Son matrices generadas por np.meshgrid() para definir la cuadrícula.
- U, V	--> Componentes del vector en las direcciones X e Y. Determinan la dirección y magnitud de cada flecha.
- color	--> Color de las flechas (opcional). Puede ser un solo color o una lista de colores.
- scale	--> Factor de escala para ajustar el tamaño de las flechas. Cuanto mayor sea, más pequeñas se verán las flechas.

````python
# Se crean las posiciones X y Y con np.meshgrid que generan matrices con las posiciones de los puntos
# de dimensiones 5x5, es decir, 25 puntos en total
# Cada np.arange genera un vector con 5 elementos
# meshgrid combina los dos vectores en matrices de 5x5. El formato de meshgrid es el siguiente: 
# X, Y = np.meshgrid([vector en x], [vector en y])
X, Y = np.meshgrid(np.arange(0, 5, 1), np.arange(0, 5, 1)) 
# Se crean los vectores U y V con las componentes de los vectores en cada punto
U = np.cos(X)
V = np.sin(Y)
# Se grafican los vectores que apuntan en la dirección de U y V
# en las posiciones dadas por X y Y 
plt.quiver(X, Y, U, V)
plt.title('Gráfico de vectores')
plt.show()
````

## <a id='toc1_19_'></a>[Diagrama de superficie 3D (plot_surface())](#toc0_)

Se representa una superficie tridimensional basada en una función matemática.     

**ax.plot_surface(X, Y, Z, cmap='plasma')**   

Parámetros   
- X, Y	--> Coordenadas de la cuadrícula en el plano XY. Se generan con np.meshgrid().
- Z	    --> Altura de la superficie en cada punto (X, Y). Es una matriz del mismo tamaño que X y Y.
- cmap	--> Mapa de colores aplicado a la superficie. Ejemplo: 'viridis', 'plasma', 'coolwarm', etc.
- edgecolor	--> Color de los bordes de la superficie. Puede ser 'black', 'white', o None para ocultarlos.

````python
# Se importa la clase Axes3D de la librería mpl_toolkits.mplot3d de matplotlib
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Se crea un subplot 3D

# Se crean los datos para la superficie
X, Y = np.meshgrid(np.linspace(-3, 3, 30), np.linspace(-3, 3, 30)) # Se crean las posiciones X y Y, usando meshgrid con 30 puntos en cada dirección 
                                                                   # usando np.linspace que nos genera un vector con 30 elementos entre -3 y 3
Z = np.sin(np.sqrt(X**2 + Y**2)) # Se crea la matriz Z con los valores de la función seno de la raíz cuadrada de X^2 + Y^2

ax.plot_surface(X, Y, Z, cmap='plasma') # Se grafica la superficie con plot_surface y el valor de cmap 'plasma' para el mapa de colores
ax.set_title('Superficie 3D')
plt.show()
````

## <a id='toc1_20_'></a>[Diagrama de Campo de flujo en 2D](#toc0_)

Se utiliza para representar el flujo de un campo vectorial continuo.

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de coordenadas
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# Definir las componentes del campo vectorial (U: dirección X, V: dirección Y)
U = -Y  # Movimiento en X depende de Y
V = X   # Movimiento en Y depende de X

# Crear el gráfico de líneas de corriente
plt.figure(figsize=(6,6))
plt.streamplot(X, Y, U, V, color='b', linewidth=1, density=1)

# Configuración del gráfico
plt.title("Gráfico de Líneas de Corriente con streamplot()")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()
````

### <a id='toc1_20_1_'></a>[Explicación del código:](#toc0_)

1. Definimos una malla de puntos (X, Y): Se usa np.meshgrid() para generar coordenadas en un rango de valores (-3 a 3 en este caso).
2. Creamos un campo vectorial (U, V):
    - U = -Y: El flujo en la dirección X depende de Y.
    - V = X: El flujo en la dirección Y depende de X.    
    Esto crea un flujo que rota en sentido antihorario, simulando un vórtice.
3. Dibujamos las líneas de corriente con plt.streamplot():
    - X, Y: Posiciones en la malla.
    - U, V: Dirección del flujo en cada punto.
    - color='b': Las líneas serán azules.
    - linewidth=1: Grosor de las líneas.
    - density=1: Ajusta la cantidad de líneas en la gráfica.
4. Agregamos etiquetas y rejilla para mejorar la visualización.

## <a id='toc1_21_'></a>[Gráfica de coordenadas polares](#toc0_)

Representación de funciones usando coordenadas polares

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir los ángulos (theta) en radianes
theta = np.linspace(0, 2 * np.pi, 100)  # De 0 a 2π (una vuelta completa)

# Definir el radio en función del ángulo
r = 1 + np.sin(4 * theta)  # Función sinusoidal con 4 oscilaciones

# Crear la figura y el eje polar
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Dibujar la gráfica polar
ax.plot(theta, r, color='r', linewidth=2)

# Configuración del gráfico
ax.set_title("Gráfico Polar de una Función Sinusoidal", va='bottom')

# Mostrar la gráfica
plt.show()
````

### <a id='toc1_21_1_'></a>[Explicación del código:](#toc0_)

1. Definimos los ángulos (theta):
    - Se usa np.linspace(0, 2 * np.pi, 100) para generar 100 puntos equidistantes entre 0 y 360° (2π radianes).
2. Definimos el radio (r):
    - r = 1 + np.sin(4 * theta): Se genera una función sinusoidal con 4 oscilaciones alrededor de un radio base de 1.
3. Creamos el gráfico polar con plt.subplots(subplot_kw={'projection': 'polar'}):
    - Se define un eje de tipo polar para dibujar la gráfica en coordenadas angulares.
4. Trazamos la curva en coordenadas polares con ax.plot(theta, r, color='r', linewidth=2).
5. Agregamos un título y configuramos el gráfico para que sea más legible.

## <a id='toc1_23_'></a>[Annotating (Anotaciones)](#toc0_)

Con ```ax.annotate('notes', xy=(xcoor, ycoor), xytext=(xcoor, ycoor), arrowprops={'arrowstyle':'->', 'color':'gray'})``` es posible poner texto sobre el gráfico en coordenadas específicas para recalcar algo.     
También es posible poner el texto sobre coordenadas específicas.     
Por último, se puede añadir una flecha que indique el texto con el punto que deseamos recalcar, y se puede personalizar la flecha dentro de un diccionario. [Aquí](https://matplotlib.org/stable/api/text_api.html) están todas las formas para personalizar la flecha.

````python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(np.sin(np.linspace(0, 10)))
plt.annotate('Mínimo',
             xy = (23, -0.95),
             xytext = (23, 0.25),
             arrowprops = dict(facecolor = 'black', width = 0.2, headwidth = 8),
             horizontalalignment = 'center')

plt.show() 
 
 
 
 
````
