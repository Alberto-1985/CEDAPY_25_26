```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1️⃣ Creamos un conjunto de datos de ejemplo (2 características)
X = np.array([
    [80, 180],
    [60, 170],
    [90, 190],
    [100, 200],
    [50, 160]
])

# 2️⃣ Creamos el escalador estándar
scaler_std = StandardScaler()  #creamos una variable de tipo StandardScaler (instanciamos)

# 3️⃣ Ajustamos y transformamos los datos
X_scaled = scaler_std.fit_transform(X) #del objeto usamos el metodo fit_transform(argumento array X), 
                                    #calcula la media y la desviacion estandar de cada columna y aplica la formula a cada valor de la fila

# 4️⃣ Mostramos los resultados
print("Datos originales:\n", X)
print("\nDatos escalados:\n", X_scaled)

# 5️⃣ Visualización
plt.figure(figsize=(8, 4))

# Antes del escalado
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='red')
plt.title("Antes del escalado")
plt.xlabel("Peso")
plt.ylabel("Altura")

# Después del escalado
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='blue')
plt.title("Después del escalado")
plt.xlabel("Peso (estandarizado)")
plt.ylabel("Altura (estandarizada)")

plt.tight_layout()
plt.show()
```
